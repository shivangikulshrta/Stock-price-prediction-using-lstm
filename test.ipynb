```python
dataset = pd.read_csv("result_dataset.csv")
dataset.head()
dataset = dataset[dataset["company_name"]=="APPLE"]
dataset = dataset.drop("company_name",axis=1)
dataset
```

```python
x = dataset.drop("Close" ,axis= 1)
x = dataset.drop("Date" ,axis= 1)
x
```

```python
y = dataset["Close"]
type(y)
```

```python


# Scale the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(x.filter(["Close"]))

scaled_data

```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size = 0.2)
```

```python

from keras.models import Sequential
from keras.layers import Dense, LSTM
# Build the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_train, y_train, batch_size=1, epochs=50)
```

```python
# Scale the target variable separately
scaler_y = MinMaxScaler(feature_range=(0, 1))
y_train_scaled = scaler_y.fit_transform(np.array(y_train).reshape(-1, 1))
y_test_scaled = scaler_y.transform(np.array(y_test).reshape(-1, 1))
```

```python
# Make predictions
predictions = model.predict(x_test)
predictions = scaler_y.inverse_transform(predictions)
```

```python
from sklearn.metrics import mean_squared_error 
# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, predictions))
print(f"RMSE: {rmse}")
```

```python
# Prepare data for visualization
data = dataset[['Date', 'Close']]
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
# Create train and validation datasets for plotting
train_len = len(y_train)
train = data.iloc[:train_len]
valid = data.iloc[train_len:]
# Assign predictions to the validation set
valid['Predictions'] = np.nan
valid.iloc[:len(predictions), valid.columns.get_loc('Predictions')] = predictions
```

```python

import matplotlib.pyplot as plt
# Plot the data
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date')
plt.ylabel('Close Price USD ($)')
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```

```python
# Inverse transform the training and validation data
train_scaled = scaler_y.inverse_transform(y_train_scaled)
valid_scaled = scaler_y.inverse_transform(np.array(valid['Close']).reshape(-1, 1))

# Update train and valid DataFrames with the inverse-transformed values
train['Close'] = train_scaled
valid['Close'] = valid_scaled

# Assign predictions to the validation set
valid['Predictions'] = np.nan
valid.iloc[:len(predictions), valid.columns.get_loc('Predictions')] = predictions

# Plot the data with inverse-transformed values
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date')
plt.ylabel('Close Price USD ($)')
plt.plot(train['Close'], label='Train')
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```
